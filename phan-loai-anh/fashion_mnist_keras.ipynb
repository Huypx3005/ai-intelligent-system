## Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense, Dropout

from random import *
```

```python
data_train = pd.read_csv('fashion-mnist_train.csv', header = 0)
data_test = pd.read_csv('fashion-mnist_test.csv', header = 0)
```

```python
data_train.head()
```

```python
X_train = np.array(data_train.drop('label', axis=1))
y_train = np.array(data_train['label'].values)
X_test = np.array(data_test.drop('label', axis=1))
y_test = np.array(data_test['label'].values)
```

```python
X_train_origin = X_train.reshape(len(X_train), 28, 28)
X_test_origin = X_test.reshape(len(X_test), 28, 28)
X_train_origin.shape
```

## Normalization

```python
# let’s normalize the input, 
# so that instead of having values from 0 to 255 (the grayscale values),
#  it has only values between 0 and 1.
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

## Reshape lại X và y để tiến hành nhân ma trận trên TensorFlow

```python
train = X_train.transpose()
labels_ = np.zeros((60000, 10))
labels_[np.arange(60000), y_train] = 1
labels_ = labels_.transpose()
labels_ = np.array(labels_)

train.shape, labels_.shape
```

```python
dev = X_test.transpose()
labels_dev_ = np.zeros((10000, 10))
labels_dev_[np.arange(10000), y_test] = 1
labels_dev_ = labels_dev_.transpose()

dev.shape, labels_dev_.shape
```

## Keras Model
Z = W * X + b

### epochs = 20

```python
model = Sequential([
    Dense(10, input_shape=(784,), activation='relu'),
    Dense(10, activation='relu'),
    Dense(10, activation='relu'),
    Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=20)
```

```python
model.evaluate(X_test, y_test)
```

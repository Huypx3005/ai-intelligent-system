# Xây dựng mô hình Machine Learning dự đoán 1 người có bị tiểu đường hay không? 
(Outcome: 1: Tiểu đường, 0: Không bị tiểu đường)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle

from sklearn import linear_model
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm

from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import MinMaxScaler
```

## Loading the data

```python
diabetes = pd.read_csv("./diabetes.csv")
```

```python
diabetes.info()
```

```python
diabetes.head()
```

## Cleaning the Data

```python
#---check for null values---
print("Nulls")
print("=====")
print(diabetes.isnull().sum())
```

```python
#---check for null values---
print("Nulls")
print("=====")
print(diabetes.isna().sum())
```

```python
#---check for 0s---
print("0s")
print("==")
print(diabetes.eq(0).sum())
```

```python
# replace the 0 values with NaN
diabetes[['Glucose','BloodPressure','SkinThickness', \
    'Insulin','BMI','DiabetesPedigreeFunction','Age']] = \
    diabetes[['Glucose','BloodPressure','SkinThickness', \
        'Insulin','BMI','DiabetesPedigreeFunction','Age']].replace \
        (0,np.NaN)
```

```python
print(diabetes.eq(0).sum())
```

```python
# replace them with the mean of each column
diabetes.fillna(diabetes.mean(numeric_only=True), inplace = True)
```

## Correlation between the features

```python
# Examining the Correlation Between the Features
corr = diabetes.corr()
print(corr)
```

```python
# Plotting the Correlation Between Features
fig, ax = plt.subplots(figsize=(10, 10))
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(diabetes.columns),1)
ax.set_xticks(ticks)
ax.set_xticklabels(diabetes.columns)
plt.xticks(rotation = 90)
ax.set_yticklabels(diabetes.columns)
ax.set_yticks(ticks)
#---print the correlation factor---
for i in range(diabetes.shape[1]):
    for j in range(9):
        text = ax.text(j, i, round(corr.iloc[i][j],2),ha="center", va="center", color="w")
plt.show()
```

## Normalization

```python
#---features---
X = diabetes.drop("Outcome", axis=1)
#---label---
y = diabetes.iloc[:,8]
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
scaled = scaler.transform(X)
X_scaled = pd.DataFrame(scaled, columns=diabetes.columns[:-1])
X_scaled
```

## Train test split

```python
# Create train and test dataset with an 80:20 split+
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=2018)

#Check the shape of each new dataset created
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

## Find algorithm has highest accuracy

```python
result = []
```

### Logistic Regression

```python
# Logistic Regression
log_regress = linear_model.LogisticRegression()
log_regress_score = cross_val_score(log_regress, X, y, cv=10, \
                                    scoring='accuracy').mean()

result.append(log_regress_score)
print(log_regress_score)
```

### K-Nearest Neighbors

```python
# K-Nearest Neighbors
#-------------------------------------
#---empty list that will hold cv (cross-validates) scores---
cv_scores = []
#---number of folds---
folds = 10
#---creating odd list of K for KNN---
ks = list(range(1,int(len(X) * ((folds - 1)/folds)), 2))
#---perform k-fold cross validation---
for k in ks:
    knn = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn, X, y, cv=folds, scoring='accuracy').mean()
    cv_scores.append(score)
#---get the maximum score---
knn_score = max(cv_scores)
#---find the optimal k that gives the highest score---
optimal_k = ks[cv_scores.index(knn_score)]
print(f"The optimal number of neighbors is {optimal_k}")

result.append(knn_score)
print(knn_score)
```

### Support Vector Machines 

```python
# Support Vector Machines 
#----------------------------
linear_svm = svm.SVC(kernel='linear')
linear_svm_score = cross_val_score(linear_svm, X, y, \
                                   cv=10, scoring='accuracy').mean()

result.append(linear_svm_score)
print(linear_svm_score)
```

### RBF kernel

```python
# RBF kernel
rbf = svm.SVC(kernel='rbf')
rbf_score = cross_val_score(rbf, X, y, cv=10, scoring='accuracy').mean()

result.append(rbf_score)
print(rbf_score)
```

## Selecting the Best Performing Algorithm

```python
algorithms = ["Logistic Regression", "K Nearest Neighbors", "SVM Linear Kernel", "SVM RBF Kernel"]
cv_mean = pd.DataFrame(result,index = algorithms)
cv_mean.columns=["Accuracy"]
cv_mean.sort_values(by="Accuracy", ascending=False)
```

```python
# Training and Saving the Model
log_regress = linear_model.LogisticRegression()
log_regress.fit(X_train, y_train)

#---save the model to disk---
filename = 'diabetes.sav'

#---write to the file using write and binary mode---
pickle.dump(log_regress, open(filename, 'wb'))

#---load the model from disk---
loaded_model = pickle.load(open(filename, 'rb'))

# Make predictions
y_pred = loaded_model.predict(X_test)
```

```python
print(f"accuracy: {accuracy_score(y_test, y_pred)}")
print(f"classification_report: {classification_report(y_test, y_pred)}")
```
